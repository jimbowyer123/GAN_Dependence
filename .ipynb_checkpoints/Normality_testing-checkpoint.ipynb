```python
sample=X[:1000,:]
```

```python
distances=np.zeros((1000,))
for i in range(1000):
    point=sample[i,:]
    #print(point)
    distances[i]=np.dot(point,point)
#print(distances)
#m_distances=np.sqrt(distances)
#print(m_distances)
ordered_distances=np.sort(distances)
print(ordered_distances)
```

```python
quantiles=np.linspace(1,1000,num=1000)
quantiles=(quantiles-0.5)/1000
chi=chi2
quants=chi.ppf(quantiles,df=100)
```

```python
plt.plot(quants,ordered_distances)
```
